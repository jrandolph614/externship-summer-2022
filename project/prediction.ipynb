```python
import pandas as pd
import numpy as np 
import joblib
from pipeline import pipeline
grid = joblib.load('models/model.pkl')
vectorizer = joblib.load('models/vectorizer.pkl')
```

```python
df = pd.read_csv('data/cleanData.csv')
df = df['clean_text'].to_frame()
df.rename(columns={'clean_text':'tweet'},inplace=True)
pipe = pipeline(df)
X = pipe.returnX(vectorizer)
df = pipe.returnDf()
y = grid.predict(X)
df['suspicion'] = y
```

```python
df[['tweet','suspicion']]
```

```python
frame = pd.read_csv('data/cleanData.csv',index_col=0)
frame.drop_duplicates(['text','date'],inplace=True)
print(len(frame))
frame = frame.merge(df[['tweet','suspicion']],how='inner',left_on='clean_text',right_on='tweet')
frame.drop_duplicates(['text','date'],inplace=True)
print(len(frame))
```

```python
frame.to_csv('data/results.csv')
```
